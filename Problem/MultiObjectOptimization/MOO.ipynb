### Multi-Objective Optimization(多目标优化, MOO)

  - 在多目标优化（Multi-Objective Optimization, MOO）中，我们不再追求一个单一“最优”解（因为多个目标常常相互冲突），而是寻找一组「无法被全面超越的解」，即 **帕累托最优解（Pareto Optimal Solutions）**。所有这些解在目标函数空间（而非策略/参数空间）中构成的集合，就叫作 **帕累托前沿（Pareto Front）**。

#### 关键概念

  1. 支配（Dominance）关系，这是判断一个解是否“优于”另一个解的核心标准：
    
  2. 非支配解（Non-dominated Solution）若一个策略 π 不被任何其他可行策略支配，则称 π 为非支配解，也就是帕累托最优解：


#### 衡量 MORL 近似 Pareto Front 好坏的指标
  - hyper volume（超体积）——用于量化一个多目标优化算法所找到的解集 P 的质量。
  - sparsity（稀疏性）——用于量化解集的多样性。
    
1. hyper volume（超体积）

  

  measure(Λ):
    我们希望找到一个统一的、严谨的数学工具 Λ，它能满足以下核心公理，并将“长度/面积/体积”的概念推广到任意维度 L 和更复杂的集合上（比如一个圆、一个分形，甚至是无法用传统方法描述的集合）。
    Lebesgue measure 是一种测度方法。

  

  > 通俗理解：hyper volume 衡量从参考点 r 到解集 P 中所有解所共同“支配”的目标空间区域的体积，越大越好。

2. sparsity（稀疏性）

  

  

  > 通俗理解：稀疏度越低越好：解集分布稠密，可供选择的解集多。
