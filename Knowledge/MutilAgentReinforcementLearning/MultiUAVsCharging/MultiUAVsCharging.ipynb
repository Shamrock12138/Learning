### Multi UAVs Charging Problem(MUCP，多无人机充电问题)

**任务场景**

多无人机的充电问题，包含两类无人机：任务机（Task UAVs）、充电机（Charging UAVs）

* 任务机负责执行任务。（不是此问题的重点）

* 充电机负责给任务机充电，保证任务机不会因为没电终止。

**决策方式**

* 任务机随机移动，电量消耗快（每步-2）

* 充电机电量充足（初始100），可给相邻任务机充电（每步-1移动）

  1. 停留在基站充电：充电机在基站静止 -> +20电量/步（上限100）
  2. 为任务机充电：每一次为任务机充电 -> -10电量/步（自身）+9电量/步（任务机）

* 终止条件：任一无人机电量≤0 或 达到最大步数

* 两种算法；

  算法一用于群体决策，为每个充电机分配任务，以及提供任务信息。
  
  算法二用于为单个充电机决策，负责具体决策此充电机的下一动作。

  训练：单个充电机，多个任务机的情况下进行训练算法二。
