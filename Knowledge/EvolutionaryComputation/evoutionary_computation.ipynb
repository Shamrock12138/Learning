### Evolutionary Computation (进化计算, EC)

#### 定义  
**进化计算（Evolutionary Computation, EC）** 是受生物进化机制（如自然选择、遗传变异）启发的一类**基于种群的随机搜索与优化方法**，用于求解复杂、非线性、黑箱优化问题，无需梯度信息。

#### 核心机制（五大要素）  
1. **初始化（Initialization）**：随机生成初始种群（一组候选解）；  
2. **评估（Evaluation）**：计算每个个体的适应度（fitness）；  
3. **选择（Selection）**：根据适应度挑选较优个体作为父代；  
4. **重组（Recombination / Crossover）**：对父代进行基因交换，生成子代；  
5. **变异（Mutation）**：以一定概率随机扰动子代基因，维持种群多样性。

#### 主要分支  
| 分支 | 英文缩写 | 特点 |
|------|---------|------|
| **遗传算法** | GA | 以二进制或实数串编码，强调选择、交叉、变异操作 |
| **进化策略** | ES | 主要用于实数优化，常自适应调整变异步长 |
| **遗传编程** | GP | 以树结构编码计算机程序，用于自动编程与符号回归 |
| **进化规划** | EP | 早期用于有限状态机演化，侧重变异与选择 |
