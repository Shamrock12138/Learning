### Multi-Objective Optimization(多目标优化, MOO)

  - 在多目标优化（Multi-Objective Optimization, MOO）中，我们不再追求一个单一“最优”解（因为多个目标常常相互冲突），而是寻找一组「无法被全面超越的解」，即 **帕累托最优解（Pareto Optimal Solutions）**。所有这些解在目标函数空间（而非策略/参数空间）中构成的集合，就叫作 **帕累托前沿（Pareto Front）**。
