### Universal Value Function Approximators (UVFA, 通用价值函数逼近器)

  * 论文：

    * 《Universal Value Function Approximators》  
      （Proceedings of the 32 nd International Conference on Machine Learning, Lille, France, 2015）

    * 《Hierarchical Universal Value Function Approximators》（2024）

  传统RL局限：
    一个任务=一个策略 - 例如agent想要将箱子推到(1, 1)，通过训练agent实现了“推到(1, 1)”，但此时如果我想让agent推到(2, 2)，我就需要重新训练。

  UVFA做法:
    一个策略通吃所有目标 - 将“目标”变成输入的一部分
